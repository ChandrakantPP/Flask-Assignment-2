Q1. Explain GET and POST methods.

Ans:- GET:- GET carries request parameter appended in URL string 
      POST:- POST carries request parameter in message body which makes it more secure way of transferring data from client to server.

Q2. Why is request used in Flask?

Ans:-  to perform both sending and receiving operations from a user's browser to the server and process the request data from the server.

Q3. Why is redirect() used in Flask?

Ans:-  to send the user to a particular URL with the status code.

Q4. What are templates in Flask? Why is the render_template() function used?

Ans:- Templates are files that contain static data as well as placeholders for dynamic data. 

## why render_template() function used:-
render_template is used to generate output from a template file based on the Jinja2 engine that is found in the application's templates folder.
